```python
# Input an integer, determine and display 'prime' if the integer is prime else display 'not prime'.
while True:
    try:
        num = int(input("Enter an integer: "))
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
    except TypeError:
        print("Cannot use float values. Please enter an integer.")

if num < 0:
    print("Negative numbers are not prime.")

if num == 1:
    print("not prime")

if num == 2:
    print("prime")

if num > 2:
    for i in range(2, num//2 + 1):
        if (num % i) == 0:
            print("not prime")
            break
        if num == num//2:
            print("prime")
```

```python
# Compute for the cube of integers.
while True:
    try:
        num = int(input("Enter an integer: "))
        break

    except ValueError:
        print("Invalid input. Please enter an integer.")

        
for i in range(1, num + 1):
    cube = i ** 3
    print("The cube of", i, "is", cube)
```

```python
# Generate a hollow square based on the + integer input side length n.
n = 0

while n < 2:
    try:
        n = int(input("Enter the side length of the square: "))
        if n <= 1:
            print("A square cannot be formed with the given side length. Please enter an integer greater than 1.")
    except TypeError:
        print("Invalid input. Please enter an integer.")
    except ValueError:
        print("Unable to use float values. Please enter an integer.")

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("+", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
# Generate an inverted right triangle of height n
n = 0

while n < 2:
    try:
        n = int(input("Enter the height of the triangle: "))
        if n <= 1:
            print("A triangle cannot be formed with the given height. Please enter an integer greater than 1.")

    except TypeError:
        print("Invalid input. Please enter an integer.")
    except ValueError:
        print("Unable to use float values. Please enter an integer.")

for i in range(n, 0, -1):
    print("* " * i)
```
