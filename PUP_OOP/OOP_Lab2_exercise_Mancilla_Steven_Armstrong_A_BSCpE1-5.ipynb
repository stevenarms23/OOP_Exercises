## Lab exercise 1 Diamond Shape:
### Write a Python function named print_diamond that takes an odd integer n as an argument and prints a diamond shape with a width of n using the * character.

    
Note: If an even number is passed, the function should return "Please provide an odd integer."


```python
def space_counter(num, star_ctr):
    return (num - star_ctr) // 2


def star_counter(num, iteration):
    return num - abs(num - iteration)


def main():
    try:
        num = int(input("Enter an odd number: ")) 
        if num % 2 == 0:
            raise ValueError
        
        for i in range(1, num*2+1, 2):
            star_ctr = star_counter(num, i)
            space_ctr = space_counter(num, star_ctr)
            print(" " * space_ctr + "*" * star_ctr)
            
    except ValueError:
        print("Please provide an odd integer.")
        main()


if __name__ == "__main__":
    main()
```

## Lab exercise 2
###	Create a program that will delete all duplicate characters in a string 
 Input starts with a number N  and is followed by N strings  
Output : Print the character with all the duplicate characters removed.

#### Sample Input #1
4  
Harrenhal  
Drogos  
Thoros of Myr  
Iron Born  

#### Sample Output #1
enl  
Drgs  
ThsfMy  
IB  


```python
def main():
    try:
        n = int(input("Enter amount of strings"))
        if n < 0:
            raise ValueError
        if n == 0:
            while True:
                should_continue = input("You tried to input 0 strings. Do you want to try again? (y/n): ").lower()
                if should_continue == 'y':
                    main()
                elif should_continue == 'n':
                    print("Thank you for using the program. Goodbye!")
                    return
        string_arr = dict()
        for i in range(1, n+1):
            input_str = input(f"Enter string #{i}: ").split(" ")
            unspaced = "".join(input_str)
            string_arr[unspaced] = "".join(char for char in unspaced if unspaced.lower().count(char.lower()) == 1)
            print(f"Unique characters in '{unspaced}': {string_arr[unspaced]}")
    except ValueError:
        print("Please enter a valid amount of strings.")
        main()
    except TypeError:
        print("Please provide a string.")
        main()

if __name__ == "__main__":
    main()
```

## Lab exercise 3
##3 Provide an implementation for a function called vowelsToUpper with the following signature:

method name : vowelsToUpper

input argument : String 

return argument : String


vowelsToUpper must return a version of its String argument with all its vowels changed to their uppercase forms. Nonvowel characters stay as is.

### Sample Calls

vowelsToUpper "" == ""

vowelsToUpper "Hello, world!" == "HEllO, wOrld!"

vowelsToUpper "hello hi bye" == "hEllO hI byE"



```python
def vowelsToUpper(string: str) -> str:
    """
    Returns a version of its string argument with all its vowels changed to their uppercase form. Nonvowel characters stay as is.
    """
    vowels = "aeiou"
    return "".join(char.upper() if char in vowels else char for char in string)

def main():
    try:
        input_str = input("Enter a string: ")
        if not isinstance(input_str, str):
            raise TypeError
        result = vowelsToUpper(input_str)
        print(f"Result: {result}")
    except TypeError:
        print("Please provide a valid string.")
        main()

if __name__ == "__main__":
    main()
```

## Lab exercise 4

from a given list of integers, create a new list that will compute the square of odd integer elements.

sample calls

[2,4,3] == [9]  
[0,0,1,1] == [1,1]


```python
def squareOddIntegers(int_list):
    """
    Returns a list of the squares of the odd integers in the input list.
    """
    return [x**2 for x in int_list if isinstance(x, int) and x % 2 != 0]


def main():
    try:
        input_str = input("Enter a list of integers (comma separated): ")
        int_list = [int(x.strip()) for x in input_str.split(",")]
        result = squareOddIntegers(int_list)
        print(f"Squares of odd integers: {result}")
    except ValueError:
        print("Please enter a valid list of integers.")
        main()


if __name__ == "__main__":
    main()
```

## Lab exercise 5

Create a new list of tuples from two given lists:

### sample calls

list1[1,2,3]  
list2["mark","alice","john]  

result: listOfTuple[(1,"mark"),(2,"alice"),(3,"john")] 

```python
def listOfTuples(list1, list2):
    """
    Returns a list of tuples
    """
    if len(list1) != len(list2):
        raise ValueError("Both lists must have the same length.")
    return list(zip(list1, list2))

def main():
    try:
        list1 = input("Enter the first list of elements (comma separated): ").split(",")
        list2 = input("Enter the second list of elements (comma separated): ").split(",")
        result = listOfTuples(list1, list2)
        print(f"Result: {result}")
    except ValueError as e:
        print(e)
        main()

if __name__ == "__main__":
    main()
```
